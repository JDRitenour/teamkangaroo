```python
weather_df = pd.read_csv('weatherRaw.csv', parse_dates=['DATE'])
```

```python
condensed_df = weather_df.groupby('DATE').agg({'DAPR':'mean', 'MDPR':'mean', 'PRCP':'mean', 'SNOW':'mean', 'SNWD':'mean'})
```

```python
condensed_df
```

```python
condensed_df = condensed_df.fillna(value = '0.00')
```

```python
condensed_df
```

```python
condensed_df.dtypes
```

```python
#update datatypes to float
condensed_df['DAPR'] = condensed_df['DAPR'].astype('float64')
condensed_df['MDPR'] = condensed_df['MDPR'].astype('float64')
condensed_df['PRCP'] = condensed_df['PRCP'].astype('float64')
condensed_df['SNOW'] = condensed_df['SNOW'].astype('float64')
condensed_df['SNWD'] = condensed_df['SNWD'].astype('float64')
```

```python
condensed_df[['SNOW', 'SNWD']].plot.bar(figsize = (5,3));
```

```python
df_weather = []
l = []
```

```python
for i in range(len(condensed_df)):
    start_date = condensed_df.index[i]
    for i in range(len(condensed_df)):
        l = []
        end_date = condensed_df.index[i]
        if start_date <= end_date:
            df = condensed_df[(condensed_df.index >= start_date) & (condensed_df.index <= end_date)]
            df = df.agg({'DAPR': 'mean', 'MDPR': 'mean', 'PRCP': 'mean', 'SNOW': 'mean', 'SNWD': 'mean'})
            l.extend([start_date, end_date, df['DAPR'], df['MDPR'], df['PRCP'], df['SNOW'], df['SNWD']])
            df_weather.append(l)
        
```

```python
df_final = pd.DataFrame(df_weather, columns=['start_date', 'end_date', 'avgDAPR', 'avgMDPR', 'avgPRCP', 'avgSNOW', 'avgSNWD'])
```
