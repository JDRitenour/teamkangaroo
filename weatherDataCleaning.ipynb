```python
import pandas as pd

import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
weather_df = pd.read_excel('weatherRaw.xlsx')
```

```python
weather_df
```

```python
weather_df.columns
```

```python
weather_df.describe()
```

```python
condensed_df = weather_df.groupby('DATE').agg({'DAPR':'mean', 'MDPR':'mean', 'PRCP':'mean', 'SNOW':'mean', 'SNWD':'mean'})
```

```python
condensed_df
```

```python
condensed_df = condensed_df.fillna(value = '0.00')
```

```python
condensed_df
```

```python
condensed_df.dtypes
```

```python
#update datatypes to float
condensed_df['DAPR'] = condensed_df['DAPR'].astype('float64')
condensed_df['MDPR'] = condensed_df['MDPR'].astype('float64')
condensed_df['SNOW'] = condensed_df['SNOW'].astype('float64')
condensed_df['SNWD'] = condensed_df['SNWD'].astype('float64')
```

```python
condensed_df.index = condensed_df.index.date
```

```python
condensed_df['DAPR'].plot.bar(figsize = (5,3), title = 'DAPR');
```

```python
condensed_df['MDPR'].plot.bar(figsize = (5,3), title = 'MDPR');
```

```python
condensed_df['PRCP'].plot.bar(figsize = (5,3), title = 'PRCP');
```

```python
condensed_df['SNOW'].plot.bar(figsize = (5,3), title = 'SNOW');
```
